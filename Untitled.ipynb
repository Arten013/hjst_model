```python
from hjst_model.config import DatasetKVSConfig, LayerModelConfig
import numpy as np
import scipy
```

```python
TESTSET = "aichi_muni"

testset_conf = DatasetKVSConfig()
testset_conf.link("./configs/dataset.conf")
testset_conf.change_section(TESTSET)
testset = testset_conf.prepare_dataset()
```

```python
layerconf = LayerModelConfig()
layerconf.link("./configs/layer.conf")
layerconf.change_section("wva_tokai_spm_16000-Sentence")
testlayer = layerconf.load_model()
```

```python
testset.set_iterator_mode(level="Sentence", tag=True, sentence=False)
```

```python
tags = [tag for tag in testset]
vectors = np.matrix([testlayer[tag] for tag in tags])
```

```python
vectors.shape
```

```python
tree = scipy.spatial.cKDTree(vectors)
```

```python
query = vectors[0]
```

```python
similarities, vids = tree.query(query, 100)
for i in range(100):
    print(i, round(similarities[0][i], 3))
    tag = tags[vids[0][i]]
    print(testset.kvsdicts["texts"]["Sentence"][tag])
```

```python
%timeit tree.sparse_distance_matrix(tree, 0.6).toarray()
```
